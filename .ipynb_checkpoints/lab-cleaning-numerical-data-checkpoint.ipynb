```python
import pandas as pd
import numpy as np
import datetime
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv') 
```

```python
data = data.rename(columns={'EmploymentStatus':'Employment Status'})
```

```python
cols = [col_name.lower().replace(' ', '_') for col_name in data.columns]
data.columns = cols
```

```python
data.columns
```

```python
corr_data = data.select_dtypes([np.number])
correlations_matrix = corr_data.corr()
```

```python
mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True 
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

# There is no high correlation among the variables (greater than 0.9).

```python
data_num = data.select_dtypes("number")
```

```python
for column in data_num.columns:
    sns.distplot(data_num[column])
    plt.show()
```

#### The variables have somehow a normal distribution, with Customer Lifetime Value, Monthly Premium Auto and Total Claim Amount having some outliers. We can say that Customer Lifetime Value mostly ranges between zero and around 20000. Monthly Premium Auto is around 50 to 150 range and Total Claim Amount lies mostly up till 1000. 

```python
data_cat = data.select_dtypes("object")
```

```python
data_cat = data_cat.drop('customer',  axis=1)
```

```python
data_cat
```

```python
for column in data_cat.columns:
    print(column)
    display(data_cat[column].value_counts())
    sns.barplot(x=data_cat[column].value_counts().index, y=data_cat[column].value_counts(), data=data_cat)
    plt.xticks(rotation = 45)
    plt.show()
```

#### The average customer is a married female who resides in California and lives in Suburban areas , has a Basic coverage 
#### contracted for her personal, four-door, medsize vehicle with a L3 Policy, has a High Education degree and is employed.
#### Most of them contact the company via agents.
